```python
def plot(path, loc="lower right"):
    train_models, val_models = os.listdir(path+"/train"), os.listdir(path+"/val")
    plt.figure(dpi=200)
    plt.gca().yaxis.grid(True, linestyle='--')
    colors = ['lightskyblue', 'red', 'black']
    i = 0
    for model in train_models:
        if ".DS_Store" not in model:
            values = np.load(path + "/train/" + model)
            plt.plot(values, label=model.replace(".npy", ""), color=colors[i])
            i+=1
        
    i = 0    
    for model in train_models:
        if ".DS_Store" not in model:
            values = np.load(path + "/val/" + model)
            plt.plot(values, color=colors[i], linestyle='--', linewidth=1)
            i += 1
            
    plt.legend(loc=loc)
    plt.ylabel(path)
    plt.xlabel("Epochs")
    plt.xticks(np.arange(0, len(values)+1, 5))
    plt.savefig(path+".png")
```

```python
plot("Accuracy")
```
