```python
import numpy as np 
import matplotlib.pyplot as plt
import os

img = plt.imread(r'cityscapes_data/train/2.jpg')
plt.figure(dpi = 150)
plt.imshow(img)
plt.show()
```

```python
left = img[:, :256, :]
right = img[:, 256:, :]
plt.figure(dpi = 150)
plt.subplot(121)
plt.imshow(left)
plt.subplot(122)
plt.imshow(right)
plt.show()
```

```python
def split(names, set_type):
    load_path = 'cityscapes_data/{}/'.format(set_type)
    save_path = 'dataset/{}/'.format(set_type)
    
    for name in names:
        img = plt.imread(load_path + name)
        plt.imsave(save_path + 'inputs/' + name, img[:, :256, :])
        plt.imsave(save_path + 'labels/' + name, img[:, 256:, :])
        yield img
```

## Train data generator

```python
names = sorted(os.listdir(r'cityscapes_data/train/'))[1:]  

gen = split(names, 'train')
for _ in range(len(names)):
    _ = next(gen)
```

## Validation data generator

```python
names = sorted(os.listdir(r'cityscapes_data/val/'))[1:]    

gen = split(names, 'val')
for _ in range(len(names)):
    _ = next(gen)
```

```python
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
kmeans = KMeans(13)
for idx in names[np.random.randint(0,len(names),16)]
  kmeans.fit(plt.imread('dataset/train/labels/{}.jpg'.format(idx)).reshape(-1,3))

import pickle
with open("kmeans.pkl", "wb") as f:
    pickle.dump(kmeans, f)
```
